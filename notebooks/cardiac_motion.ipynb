## Biventricular Motion

```python

from biv_lite.biv_frames import BivFrames
from biv_lite.app.ui import BivMotionUI
import ipywidgets as widgets
```

```python
def time_controls(callback, tmin, tmax, step=1):

    def set_time(change):
        step = change["new"]
        if step < 0:
            step = 0
        if step >= tmax:
            step = tmax - 1
        callback(step)

    play = widgets.Play(
        value=tmin,
        min=tmin,
        max=tmax,
        step=step,
        description="Time Step",
    )
    play.observe(set_time, "value")

    slider = widgets.IntSlider(min=tmin, max=tmax, step=1, continuous_update=True)
    widgets.jslink((play, "value"), (slider, "value"))
    return widgets.HBox([play, slider])
```

```python
biv_frames = BivFrames.from_folder('../tests/sample_frames/')
```

```python
ui = BivMotionUI(biv_frames)
```

```python
def update_mesh(frame):
    ui.replace_mesh_callback(frame)
    ui.plotter.render()
```

```python
pw = time_controls(update_mesh, tmin=0, tmax=len(ui.biv_frames)-1)
```
